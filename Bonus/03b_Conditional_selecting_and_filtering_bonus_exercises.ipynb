Read the csv file from this url:
'https://drive.google.com/file/d/1P2EyVCFAPeJcQ_lAN2W-bjwtbizOeFXE/view?usp=drive_link'

Name it `sets_df`

```python
# Your code here
url = 'https://drive.google.com/file/d/1P2EyVCFAPeJcQ_lAN2W-bjwtbizOeFXE/view?usp=drive_link'
path = 'https://drive.google.com/uc?export=download&id=' + url.split('/')[-2]
sets_df = pd.read_csv(path)
sets_df
```

#### Exercise 1.  
Find and display all sets from the year 2005.

```python
# Your code here
sets_df.loc[sets_df['year'] == 2005]
```

#### Exercise 2
Find and display sets from year 2010 with more than 500 parts.

```python
# Your code here
sets_df.loc[(sets_df['year'] == 2010) & (sets_df['num_parts'] > 500)]
```

#### Exercise 3
Find and display sets having between 100 and 200 parts by using `.between()`.

```python
# Your code here
sets_df.loc[sets_df['num_parts'].between(100, 200)]
```

#### Exercise 4
Find and display sets that are **not** from the year 2002.

```python
# Your code here
sets_df.loc[~(sets_df['year'] == 2000)]
```

#### Exercise 5
Display all sets with `theme_id` 158

#### Exercise 5
Find and display the 10 biggest sets (by number of parts) by using `.nlargest()`.  


```python
# Your code here
sets_df.nlargest(10, 'num_parts')
```

#### Exercise 6

#### Exercise 9
Find the 5 smallest sets from the 1990's.


```python
# Your code here
sets_df.query('year == 1990').nsmallest(5, ['num_parts'])

sets_mask = sets_df['year'] == 1990
sets_df.loc[sets_mask].nsmallest(5, ['num_parts'])
```

#### Exercise 10
Find and display sets with more than 100 parts from the years 2010-2015.   


```python
# Your code here
sets_num_part_mask = sets_df['num_parts'] > 100
sets_year_mask = sets_df['year'].between(2010, 2015)
sets_df.loc[sets_num_part_mask & sets_year_mask]
```
